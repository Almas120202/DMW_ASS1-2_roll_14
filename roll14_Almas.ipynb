# Preprocessing Techniques

# Data Cleaning

```python
#remove the unwanted coloumns from training and testing dataset
titanic_train=titanic_train.drop(['Cabin','Name','Ticket','PassengerId','Embarked'],1)
titanic_test=titanic_test.drop(['Cabin','Name','Ticket','PassengerId','Embarked'],1)
```

```python
#after removing unwanted columns displaying training dataset
titanic_train
```

```python
#after removing unwanted columns displaying testing dataset
titanic_test
```

```python
#removing duplicates from training dataset
titanic_train=titanic_train.drop_duplicates()
```

```python
#removing duplicates from testing dataset
titanic_test=titanic_test.drop_duplicates()
```

```python
#total no. of duplicates in train
sum(titanic_train.duplicated())
```

```python
#total no. of duplicates in test
sum(titanic_test.duplicated())
```
